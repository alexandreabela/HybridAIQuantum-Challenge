```python
import networkx as nx
import matplotlib.pyplot as plt

from utils import MNIST_partial, MNIST_partial_graph

val_dataset = MNIST_partial(data='data', split='val')
val_dataset_graph = MNIST_partial_graph(data='data', split='val')
```

```python
def display_graph(G):
    # Create a dictionary of positions where keys are nodes (i,j) and values are (x,y) coordinates
    pos = {(i,j): (j, -i) for i,j in G.nodes()}  # -i to flip vertical axis
    
    plt.figure(figsize=(4, 4))
    nx.draw(G, 
           pos=pos,           # Use our custom position mapping
           node_color=[G.nodes[node]['x'] for node in G.nodes()],
           cmap=plt.cm.gray,
           node_size=20,
           width=0.1,
           with_labels=False)
    plt.show()
```

```python
# Get a single image from your dataset
img, label = val_dataset[1]

print(f"Label is : {label}")

plt.figure()
plt.imshow(img[0], cmap="gray")
plt.axis('off')
plt.show()

grph, label_grph = val_dataset_graph.get_graphNX(1)

print(f"Label is : {label_grph}")
display_graph(grph)
```
